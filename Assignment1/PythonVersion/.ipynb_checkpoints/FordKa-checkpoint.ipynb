### Preprocessing

```python
#Create dummies.  You may want to do df.drop('MaritalStatus',axis=1)
#axis=1 means columns
df = pd.concat([df,pd.get_dummies(df['MaritalStatus'],prefix='MaritalStatus')],axis=1)
```

```python
#replaces the '2' with 0
df['Gender'] = df['Gender'].map({1:1,2:0})
```

```python
df.head().T
```

```python
#Reduce lists of questions
reduced_qs = ['Q2','Q14','Q17','Q31','Q41','Q44']
demos = ['Gender','Age','MaritalStatus_1','MaritalStatus_2','MaritalStatus_3','NumberChildren','IncomeCategory']
df_reduced_qs = df[reduced_qs]
df_demos = df[demos]
```

```python
#Scaling the demographics information.
min_max_scaler = preprocessing.MinMaxScaler()
df_demos = min_max_scaler.fit_transform(df_demos)
```

### K Means

```python
#For Visualization of the clusters

tsne = TSNE(n_components=2, random_state=0)  
df_tsne = tsne.fit_transform(df)
```

```python
plt.scatter(df_tsne[:, 0], df_tsne[:, 1])
plt.axis('equal')
```

```python
#For Visualization of the clusters

tsne = TSNE(n_components=2, random_state=0)  
df_reduced_qs_tsne = tsne.fit_transform(df_reduced_qs)
```

```python
plt.scatter(df_reduced_qs_tsne[:, 0], df_reduced_qs_tsne[:, 1])
plt.axis('equal')
```

```python
print(tsne.kl_divergence_ ) #amount of variation explained 
```

```python
#Finding the optimum number of clusters for k-means classification
wcss = []

for i in range(1, 8):
    kmeans = KMeans(n_clusters = i, max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(df_reduced_qs_tsne)
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1, 8), wcss)
plt.title('Scree Plot')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, n_init=1000, random_state=0)
kmeans.fit(df_reduced_qs_tsne)

kmeans_cluster_assignments = kmeans.predict(df_reduced_qs_tsne)

plt.scatter(df_reduced_qs_tsne[:, 0], df_reduced_qs_tsne[:, 1], #[all rows:, in column 0]
            c=kmeans_cluster_assignments, cmap='Spectral')
# - `c` specifies how the different points should be colored (here, I set it to the
#   cluster assignments so different clusters get different colors)
plt.axis('equal')
```

```python
#Concatenates the columns with the mean scores for that cluster
np.c_[df_reduced_qs.columns,df_reduced_qs[kmeans_cluster_assignments == 0].mean(axis=0)]
```

```python
#Mapping back cluster assignments

df['Cluster'] = kmeans_cluster_assignments.tolist()
```
